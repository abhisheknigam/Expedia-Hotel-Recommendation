```python
from math import*
#from collections import Counter as mset
def jaccard_similarity(x,y):
 
    intersection = len(set.intersection(*[set(x), set(y)]))
    print ("Intersection cardinality: " + repr(intersection))
    union = len(set.union(*[set(x), set(y)]))
    print ("Union cardinality: "+ repr(union))
    return intersection/float(union)
 
print (jaccard_similarity([0,2,3,5,6,8],[1,2,6,5,7,9]))
```

```python
import operator
def getNeighbors(trainingSet, testInstance, k):
    similarities = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        similarity = jaccard_similarity(testInstance, list(trainingSet[x][1:14]))
        similarities.append((trainingSet[x], 1-similarity))
    similarities.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(similarities[x][0])
    return neighbors
```

```python
def getResponse(neighbors):
	classVotes = {}
	for x in range(len(neighbors)):
		response = neighbors[x][-1]
		if response in classVotes:
			classVotes[response] += 1
		else:
			classVotes[response] = 1
	sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
	return sortedVotes[0][0]
```

```python
from collections import Counter
predictions=[]
k = 5
#k = round(sqrt(8000))
correct = 0
for x in range(len(testSet)):
    similarities = []
    for y in range(len(trainingSet)):
        similarity = jaccard_similarity(list(testSet[x][0:14]), list(trainingSet[y][0:14]))
        similarities.append((trainingSet[y][-1], 1-similarity))
    similarities.sort(key=operator.itemgetter(1))
    classes = similarities[1:k][0]
    Votes = Counter()
    for hotel in classes:
        Votes[hotel] += 1
    prediction = (Counter(Votes).most_common(1))[0][0]
    predictions.append(prediction)
    print(repr(x)+'> predicted=' + repr(prediction) + ', actual=' + repr(testSet[x][-1]))
    if testSet[x][-1] == prediction:
        correct += 1
accuracy = (correct/float(len(testSet))) * 100.0
print('Accuracy: ' + repr(accuracy) + '%')
```
